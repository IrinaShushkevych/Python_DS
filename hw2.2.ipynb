```python
import pandas as pd
```

```python
data = pd.read_csv('2017_jun_final.csv')
data.head(5)
```

```python
size = data.shape
print(f'{size[0]} rows, {size[1]} cols')
```

```python
data.dtypes
```

```python
data.isnull().sum().to_frame('nulls')
```

```python
cols = data.loc[:, data.isnull().any()].copy().columns.tolist()
cols.remove('Язык.программирования')
print(cols)

new_data = data.drop(cols, axis=1)
print(new_data.shape)
print(new_data.isnull().sum().to_frame('nulls'))
```

```python
new_data = new_data.dropna()
```

```python
new_data.shape
```

```python
python_data = new_data[new_data['Язык.программирования'].eq('Python')]
```

```python
python_data.shape
```

```python
group_data = python_data.groupby('Должность')
```

```python
group_data['Зарплата.в.месяц'].agg(['min', 'max'])
```

```python
def fill_avg_salary(row):
    gd = python_data[python_data['Должность'].eq(row['Должность'])].groupby('Должность')
    avg = gd['Зарплата.в.месяц'].agg('mean').round(2).tolist()[0]
    return avg
```

```python
python_data.loc[:, 'avg'] = python_data.apply(fill_avg_salary, axis=1)
python_data.head(3)
```

```python
stat = python_data['avg'].describe()
stat
```

```python
stat.to_csv("average_salay_stat.csv")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = python_data[['Должность', 'avg']].drop_duplicates()
plt.bar(data['Должность'], data['avg'])
plt.title('Середня зарплата (Python)', color='blue')
plt.xlabel('Посада', fontsize='small')
plt.ylabel('Середня зарплата', fontsize='small')
plt.xticks(rotation=60)
plt.grid(axis='y')
plt.show()
```

```python
gd = python_data[['Должность', 'Возраст']].groupby('Должность')
fig, axs = plt.subplots()
for key, values in gd:
    d = values.groupby('Возраст').size()
    axs.plot(d.index.values, d.values, marker='o', label=key)

plt.title('Кількість працівників в вікових групах', y=1.1)
plt.xlabel('Вік', fontsize='small')
plt.ylabel('Кількість', fontsize='small')
plt.legend()
plt.grid()
plt.show()
```

```python
gd = python_data[python_data['Еще.студент'].eq(True)][['Должность','Еще.студент']].groupby('Должность').count().reset_index()

plt.pie(gd['Еще.студент'].values, 
        labels=gd['Должность'].values, 
        shadow=True, 
        autopct="%.0f",
        pctdistance=1.1,
        labeldistance=1.3
       )
plt.title('Кількість студентів на посадах', fontsize=15, y=1.1)
plt.show()
```
